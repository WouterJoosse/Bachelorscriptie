```python
match_insert_query = Template("INSERT INTO vectortree_results (target_id, candidate_id, threshold, distance, registration_type) VALUES ($target_id, $candidate_id, $threshold, $distance, $registration_type)")

def db_insert_match(cursor, prepared_query, row, threshold, registration_type):
    values = {
        'target_id': int(row['id_1']),
        'candidate_id': int(row['id_2']),
        'threshold': threshold,
        'distance': int(row['dist']),
        'registration_type': registration_type
    }
    cursor.execute(prepared_query.safe_substitute(values))
    
```
