```python
# figuur 4.1 - Aantal registraties per label

ylim= (0, lvl_5_dist_5[['dist','Label']].groupby('Label').count().max().values[0] *1.10)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,5), subplot_kw={'ylim': ylim})
fig.suptitle("Distribution of error over names")



# De bars - Lvl 3
ax = lvl_3_dist_3[['dist','Label']].groupby('Label').count().plot(kind='bar', title="threshold = 3", legend=False, ax=axes[0])
ax.set_xlabel("Distribution for matches with a distance of 3")
ax.set_ylabel("Number of matches")
totals = []
for p in ax.patches:
    totals.append(p.get_height())
total = sum(totals)
for p in ax.patches:
    ax.text(p.get_x() + .025, p.get_height() + max(total * 0.1, min(p.get_height() * 1.1, 500)), \
           str(round((p.get_height() / float(total)) * 100, 2)) + '%', fontsize=15, color='dimgray')


# De bars - Lvl 4
ax = lvl_4_dist_4[['dist','Label']].groupby('Label').count().plot(kind='bar', title="threshold = 4", legend=False, ax=axes[1])
ax.set_xlabel("Distribution for matched with a distance of 4")
totals = []
for p in ax.patches:
    totals.append(p.get_height())
total = sum(totals)
for p in ax.patches:
    ax.text(p.get_x() - p.get_x() * .05, p.get_height() + max(total * 0.05, min(p.get_height() * 1.1, 500)), \
           str(round((p.get_height() / float(total)) * 100, 2)) + '%', fontsize=15, color='dimgray')

# De bars - Lvl 5
ax = lvl_5_dist_5[['dist','Label']].groupby('Label').count().plot(kind='bar', title="threshold = 5", legend=False, ax=axes[2])
ax.set_xlabel("Distribution for matches with a distance of 5")
totals = []
for p in ax.patches:
    totals.append(p.get_height())
total = sum(totals)
for p in ax.patches:
    ax.text(p.get_x() - p.get_x() * .05, p.get_height() + max(total * 0.0125, min(p.get_height() * 1.1, 500)), \
           str(round((p.get_height() / float(total)) * 100, 2)) + '%', fontsize=15, color='dimgray')


if SAVEFIGS:
    fig.savefig('reports/figures/distribution_max_error_over_names_all.png', bbox_inches='tight')
plt.show()
```
