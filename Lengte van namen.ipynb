```python
hw_lvl_3 = pd.read_csv('datafiles/dataframes/working/huwelijk_lvl_3_extended.csv', sep='|', index_col=False)
```

```python
SAVEFIG = False
name_length_plot = pd.DataFrame()
name_length_plot['male_first_name'] = pd.Series(hw_lvl_3.target_name_1).map(len)
name_length_plot['male_fam_name'] = pd.Series(hw_lvl_3.target_name_2).map(len)
name_length_plot['female_first_name'] = pd.Series(hw_lvl_3.target_name_3).map(len)
name_length_plot['female_fam_name'] = pd.Series(hw_lvl_3.target_name_4).map(len)

ax = name_length_plot.plot.box(vert=False, color=dict(boxes="DarkGreen"), title="Marriage registrations - Name length")
ax.set_yticklabels(['Male first name', 'Male family name', 'Female first name', 'Female family name'])
fig = ax.get_figure()
if SAVEFIG:
    fig.savefig('reports/figures/avg_name_length_boxplot.png', bbox_inches='tight')
plt.show()
```
