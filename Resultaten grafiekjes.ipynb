```python
hw_lvl_3 = pd.read_csv('datafiles/dataframes/working/huwelijk_lvl_3_extended_2.csv', sep='|', index_col=False)
hw_lvl_4 = pd.read_csv('datafiles/dataframes/working/huwelijk_lvl_4_extended_2.csv', sep='|', index_col=False)
hw_lvl_5 = pd.read_csv('datafiles/dataframes/working/huwelijk_lvl_5_extended_2.csv', sep='|', index_col=False)
```

```python
hw_lvl_3['th'] = 3
hw_lvl_4['th'] = 4
hw_lvl_5['th'] = 5
```

```python
plotframe = pd.concat([hw_lvl_3, hw_lvl_4, hw_lvl_5])
plotframe.sort_values(['id_1', 'th'])
```

```python
plotframe_groups = plotframe[['th', 'id_1', 'id_2']].groupby(['id_1', 'th'])
```

```python
counts = plotframe_groups.count()
```

```python
counts
```

```python
totals = counts.sum(level=0)
totals
```

```python
counts1 = counts.unstack(level=1)
counts1.columns = counts1.columns.droplevel(level=0)
counts1
```

```python
counts1 = counts1.fillna(0)
```

```python
counts1[3].ravel().max()
```

```python
num_bins = np.max([counts1[3].max(), counts1[4].max(), counts1[5].max()]).astype(int) + 1
bins = np.linspace(0, 10, 11)
plt.hist(counts1[5].ravel(), bins=bins, normed=True, histtype='step')
plt.hist(counts1[4].ravel(), bins=bins, normed=True, histtype='step')
plt.hist(counts1[3].ravel(), bins=bins, normed=True, histtype='step')

plt.legend(pos="best")


plt.show()
```

```python
fig, ax = plt.subplots()
a_heights, a_bins = np.histogram(counts1[3], bins=np.linspace(0,10,11))
b_heights, b_bins = np.histogram(counts1[4], bins=a_bins)
c_heights, c_bins = np.histogram(counts1[5], bins=a_bins)

width = (a_bins[1] - a_bins[0]) / 4

ax.bar(a_bins[:-1], a_heights, width=width)
ax.bar(a_bins[:-1]+width, b_heights, width=width)
ax.bar(a_bins[:-1]+(2*width), c_heights, width=width)
plt.show()
```

```python
ed_3_hw = hw_lvl_3.drop(hw_lvl_3[(hw_lvl_3.ED_Label == '3') & (hw_lvl_3.dist_name_3 == 1)].index)[hw_lvl_3.ED_Label == '3'][['dist_name_1', 'dist_name_2', 'dist_name_3', 'dist_name_4', 'id_1']]
ed_3_hw_grouped = ed_3_hw.groupby(['dist_name_1', 'dist_name_2', 'dist_name_3', 'dist_name_4'])

ed_4_hw = hw_lvl_4[hw_lvl_4.ED_Label == '4'][['dist_name_1', 'dist_name_2', 'dist_name_3', 'dist_name_4', 'id_1']]
ed_4_hw = ed_4_hw.drop(ed_4_hw[((ed_4_hw.dist_name_1 < 4) & (ed_4_hw.dist_name_1 > 0)) |
                    ((ed_4_hw.dist_name_2 < 4) & (ed_4_hw.dist_name_2 > 0)) |
                    ((ed_4_hw.dist_name_3 < 4) & (ed_4_hw.dist_name_3 > 0)) |
                    ((ed_4_hw.dist_name_4 < 4) & (ed_4_hw.dist_name_4 > 0))].index)
ed_4_hw_grouped = ed_4_hw.groupby(['dist_name_1', 'dist_name_2', 'dist_name_3', 'dist_name_4'])

ed_5_hw = hw_lvl_5[hw_lvl_5.ED_Label == '5'][['dist_name_1', 'dist_name_2', 'dist_name_3', 'dist_name_4', 'id_1']]
ed_5_hw = ed_5_hw.drop(ed_5_hw[((ed_5_hw.dist_name_1 < 5) & (ed_5_hw.dist_name_1 > 0)) |
                    ((ed_5_hw.dist_name_2 < 5) & (ed_5_hw.dist_name_2 > 0)) |
                    ((ed_5_hw.dist_name_3 < 5) & (ed_5_hw.dist_name_3 > 0)) |
                    ((ed_5_hw.dist_name_4 < 5) & (ed_5_hw.dist_name_4 > 0))].index)
ed_5_hw_grouped = ed_5_hw.groupby(['dist_name_1', 'dist_name_2', 'dist_name_3', 'dist_name_4'])
```
