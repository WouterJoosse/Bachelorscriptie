```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from programs.utils.analyse_functies import *
SAVEFIGS = False
```

```python
matching_type = 'huwelijk'
lvl_3_samples = pd.read_csv('datafiles/dataframes/working/' + matching_type + '_lvl_3_dist_3_extended_2.csv', sep="|", index_col=False)
lvl_4_samples = pd.read_csv('datafiles/dataframes/working/' + matching_type + '_lvl_4_dist_4_extended_2.csv', sep="|", index_col=False)
lvl_5_samples = pd.read_csv('datafiles/dataframes/working/' + matching_type + '_lvl_5_dist_5_extended_2.csv', sep="|", index_col=False)
```

```python
lvl_5_samples.Label.isnull().sum()
```

```python
lvl_4_samples = lvl_4_samples[(lvl_4_samples.Label != '4')]
lvl_5_samples = lvl_5_samples[(lvl_5_samples.Label != '5') & (lvl_5_samples.Label != '2,3')]
```

```python
lvl_3_samples['RuleSet_loose'] = lvl_3_samples.apply(filter_namepairs, axis=1, args=(False,))
lvl_3_samples.groupby(['Label', 'RuleSet_loose']).count()
```

```python
lvl_3_samples
```

```python
lvl_4_samples['RuleSet_loose'] = lvl_4_samples.apply(filter_namepairs, axis=1, args=(False,))
lvl_4_samples.groupby(['Label', 'RuleSet_loose']).count()
```

```python
lvl_5_samples['RuleSet_loose'] = lvl_5_samples.apply(filter_namepairs, axis=1, args=(False,))
lvl_5_samples.groupby(['Label', 'RuleSet_loose']).count()
```

```python
lvl_3_samples['RuleSet_strict'] = lvl_3_samples.apply(filter_namepairs, axis=1, args=(True,))
lvl_3_samples.groupby(['Label', 'RuleSet_strict']).count()
```

```python
lvl_4_samples['RuleSet_strict'] = lvl_4_samples.apply(filter_namepairs, axis=1, args=(True,))
lvl_4_samples.groupby(['Label', 'RuleSet_strict']).count()
```

```python
lvl_5_samples['RuleSet_strict'] = lvl_5_samples.apply(filter_namepairs, axis=1, args=(True,))
lvl_5_samples.groupby(['Label', 'RuleSet_strict']).count()
```

```python
# Plot de resultaten in 3 plots: Per max Edit Distance. 
ylim = (0, pd.pivot_table(lvl_3_samples[lvl_3_samples.RuleSet_loose], values='id_1', index='Label', columns='RuleSet_loose', aggfunc='count').values.max()*1.02)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,5), subplot_kw={'ylim': ylim})

ax_3 = pd.pivot_table(lvl_3_samples[lvl_3_samples.RuleSet_loose], values='id_1', index='Label', columns='RuleSet_loose', aggfunc='count').plot.bar(ax=axes[0], title="Edit distance: 3")
pd.pivot_table(lvl_3_samples[lvl_3_samples.RuleSet_strict], values='id_1', index='Label', columns='RuleSet_strict', aggfunc='count').plot.bar(ax=ax_3, color='orange')

ax_4 = pd.pivot_table(lvl_4_samples[lvl_4_samples.RuleSet_loose], values='id_1', index='Label', columns='RuleSet_loose', aggfunc='count').plot.bar(ax=axes[1], title="Edit distance: 4")
pd.pivot_table(lvl_4_samples[lvl_4_samples.RuleSet_strict], values='id_1', index='Label', columns='RuleSet_strict', aggfunc='count').plot.bar(ax=ax_4, color='orange')

ax_5 = pd.pivot_table(lvl_5_samples[lvl_5_samples.RuleSet_loose], values='id_1', index='Label', columns='RuleSet_loose', aggfunc='count').plot.bar(ax=axes[2], title="Edit distance: 5")
pd.pivot_table(lvl_5_samples[lvl_5_samples.RuleSet_strict], values='id_1', index='Label', columns='RuleSet_strict', aggfunc='count').plot.bar(ax=ax_5, color='orange')

ax_3.legend_.set_visible(False)
ax_4.legend_.set_visible(False)
ax_5.legend_.set_visible(False)

fig.suptitle("Number of accepted matches", fontsize=16)
plt.savefig('reports/figures/number_accepted_matches_ruleset.png', bbox_inches='tight')
plt.show()
```
